#### Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

#### Reading Data

```python
data = pd.read_csv("/Users/krishna/Desktop/VS Code/ex1data1.txt",names=['Population','Profit'])
m=len(data)
x=np.append(np.ones((m,1)),np.array(data["Population"]).reshape((m,1)),axis=1)
y=np.array(data["Profit"]).reshape(m,1)
```

#### Cost Function

```python
def cost_function(x,y,theta):
    a=theta[0][0]
    b=theta[1][0]
    cost=0
    for i in range(m):
        cost+=(a+b*x[i][1]-y[i])**2
    return cost/(2*m)
```

#### Gradient Descent

```python
def gradient_descent(x,y,theta,alpha,iterations):
    w=[0]
    J_History=[(cost_function(x,y,theta))]
    for p in range(iterations):
        sum=0
        for i in range(m):
            sum+=((theta[0][0]+theta[1][0]*x[i][1]-y[i][0])*x[i][0])
        theta[0][0]-=(alpha*sum)/m
        sum2=0
        for i in range(m):
            sum2+=((theta[0][0]+theta[1][0]*x[i][1]-y[i][0])*x[i][1])
        theta[1][0]-=(alpha*sum2)/m
        J_History.append(cost_function(x,y,theta))
        w.append(p+1)
    return theta, J_History, w

theta, J_History, w =gradient_descent(x,y,np.zeros((2,1)),0.001,15000)
```

#### Linear Regression Plot

```python
plt.plot(x[:,1],y,'rx',label='Training Data')
plt.plot(x[:,1],x.dot(theta),label='Linear Regression')
```
